Question 1: What is Simple Linear Regression?

ans-  It predicts the value of a dependent variable (Y) based on the value of an independent variable (X) by fitting a straight line through the data points. The relationship is assumed to be linear ‚Äî meaning changes in X result in proportional changes in Y.

üßÆ Mathematical Formula
The equation of the regression line is:
\hat{Y} = b_0 + b_1X
- \hat{Y}: Predicted value of the dependent variable
- b_0: Intercept (value of Y when X = 0)
- b_1: Slope (change in Y for a one-unit change in X)
- X: Independent variable

üîç Example
Suppose you're analyzing how the number of hours studied affects exam scores:
- X: Hours studied
- Y: Exam score
- A simple linear regression might reveal that each additional hour of study increases the score by 5 points.

Question 2: What are the key assumptions of Simple Linear Regression?

ans- 1. Linearity
- The relationship between the independent variable X and the dependent variable Y is linear.
- This means the change in Y is proportional to the change in X.
2. Independence of Errors
- The residuals (errors) should be independent of each other.
- No autocorrelation ‚Äî especially important in time series data.
3. Homoscedasticity
- The variance of residuals should be constant across all levels of X.
- If residuals fan out or funnel in, it indicates heteroscedasticity, which violates this assumption.
4. Normality of Residuals
- The residuals should be approximately normally distributed.
- This is especially important for hypothesis testing and confidence intervals.
5. No Multicollinearity (only relevant in multiple regression)
- For Simple Linear Regression, this isn‚Äôt a concern since there‚Äôs only one predictor.
- But good to keep in mind when you expand to multiple regression models.
6. No Significant Outliers
- Outliers can distort the regression line and affect the slope and intercept.
- It‚Äôs important to detect and handle them appropriately.

Question 3: What is heteroscedasticity, and why is it important to address in regression models?

ans- In regression analysis, heteroscedasticity refers to a condition where the variance of the residuals (errors) is not constant across all levels of the independent variable(s). Instead of a uniform scatter of residuals, you might see a pattern ‚Äî like a funnel shape ‚Äî where errors grow larger or smaller as the predicted values increase.
This violates the assumption of homoscedasticity, which is essential for Ordinary Least Squares (OLS) regression to produce valid and efficient estimates.

‚ö†Ô∏è Why Is It Important to Address?
Ignoring heteroscedasticity can lead to:
- Biased standard errors: This affects confidence intervals and hypothesis tests.
- Misleading significance tests: You might incorrectly conclude that a variable is statistically significant.
- Inefficient estimates: OLS no longer provides the Best Linear Unbiased Estimator (BLUE).
- Poor model interpretation: Predictions may be less reliable, especially for values at the extremes.

Question 4: What is Multiple Linear Regression?

ans- Multiple Linear Regression estimates how multiple predictors (independent variables) together influence a single response (dependent variable). It fits a linear equation to the data that accounts for the contribution of each variable while holding others constant.

üßÆ Mathematical Formula
\hat{Y} = b_0 + b_1X_1 + b_2X_2 + \dots + b_nX_n
- \hat{Y}: Predicted value of the dependent variable
- b_0: Intercept
- b_1, b_2, \dots, b_n: Coefficients for each independent variable
- X_1, X_2, \dots, X_n: Independent variables

üìä Example
Imagine you're predicting house prices based on:
- Square footage
- Number of bedrooms
- Distance to city center

Question 5: What is polynomial regression, and how does it differ from linear
regression?

ans- Polynomial Regression is a flexible extension of linear regression that allows you to model non-linear relationships between the independent and dependent variables. While linear regression fits a straight line, polynomial regression fits a curve by including higher-degree terms of the predictor variable.

üî¢  Formula
Polynomial regression models the relationship using an equation like:
\hat{Y} = a_0 + a_1X + a_2X^2 + a_3X^3 + \dots + a_nX^n
- \hat{Y}: Predicted value
- X: Independent variable
- a_0, a_1, \dots, a_n: Coefficients
- n: Degree of the polynomial



üîç Key Differences from Linear Regression

Linear regression assumes a straight-line relationship between the independent and dependent variables. It‚Äôs ideal when the data points follow a consistent upward or downward trend.

Polynomial regression, on the other hand, is designed for curved relationships. It can capture more complex patterns like U-shapes, S-curves, or waves by including higher powers of the independent variable.

üîπ Model Complexity
Linear regression is simple and easy to interpret. You‚Äôre working with just one slope and one intercept.
Polynomial regression adds complexity by introducing squared, cubic, or higher-order terms. This makes the model more flexible but also harder to interpret.

üîπ Flexibility
Linear regression is limited to modeling straight-line trends. If your data bends or twists, it won‚Äôt capture that well.
Polynomial regression is much more flexible. It can fit curves and adapt to intricate data patterns ‚Äî but that flexibility comes with a risk of overfitting if the degree is too high.

üîπ Use Cases
Use linear regression when your data shows a clear linear trend ‚Äî like predicting salary based on years of experience.
Use polynomial regression when the relationship is non-linear ‚Äî like modeling population growth, where the rate accelerates over time.

üîπ Overfitting Risk
Linear regression is less prone to overfitting because of its simplicity.
Polynomial regression, especially with high-degree polynomials, can overfit the training data and perform poorly on new data.

 Example
Imagine you're modeling the growth of a plant over time:
- Linear regression might suggest a constant growth rate.
- Polynomial regression can capture acceleration or deceleration in growth, like a curve that flattens or steepens over time.







```python
#Question 6: Implement a Python program to fit a Simple Linear Regression model to the following sample data:
# X = [1, 2, 3, 4, 5]
# Y = [2.1, 4.3, 6.1, 7.9, 10.2]
#Plot the regression line over the data points.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape for sklearn
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Plotting
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Question 7: Fit a Multiple Linear Regression model on this sample data:
#Area = [1200, 1500, 1800, 2000]
# Rooms = [2, 3, 3, 4]
# Price = [250000, 300000, 320000, 370000]
#Check for multicollinearity using VIF and report the results.
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Sample data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}

df = pd.DataFrame(data)

# Independent variables
X = df[['Area', 'Rooms']]
y = df['Price']

# Fit Multiple Linear Regression
model = LinearRegression()
model.fit(X, y)

# Print coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Add constant for VIF calculation
X_const = add_constant(X)

# Calculate VIF
vif_data = pd.DataFrame()
vif_data['Feature'] = X_const.columns
vif_data['VIF'] = [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]

print("\nVariance Inflation Factor (VIF):")
print(vif_data)

```

 Interpretation of VIF
- VIF < 5: Generally acceptable
- VIF between 5‚Äì10: Moderate multicollinearity
- VIF > 10: High multicollinearity ‚Äî problematic
In this case, both Area and Rooms have VIFs around 5.83, which suggests moderate multicollinearity. It‚Äôs not critical, but worth monitoring ‚Äî especially if you expand the model with more features.


```python
#Question 8: Implement polynomial regression on the following data:
# X = [1, 2, 3, 4, 5]

# Y = [2.2, 4.8, 7.5, 11.2, 14.7]
#Fit a 2nd-degree polynomial and plot the resulting curve
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Transform features to 2nd-degree polynomial
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict values
X_range = np.linspace(1, 5, 100).reshape(-1, 1)  # Smooth curve
X_range_poly = poly.transform(X_range)
Y_pred = model.predict(X_range_poly)

# Plotting
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X_range, Y_pred, color='red', label='Polynomial Curve')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2nd-Degree Polynomial Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Question 9: Create a residuals plot for a regression model trained on this data:
# X = [10, 20, 30, 40, 50]
# Y = [15, 35, 40, 50, 65]
#Assess heteroscedasticity by examining the spread of residuals.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit the model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Plot residuals
plt.scatter(Y_pred, residuals, color='purple', edgecolor='black')
plt.axhline(y=0, color='gray', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.grid(True)
plt.show()
```

- Variance Inflation Factor (VIF): Calculate VIF for each feature. Values > 5 or 10 suggest problematic correlation.
- Correlation matrix: Check for high pairwise correlations between predictors.

- Switch to models that adjust standard errors, like Huber regression or OLS with robust standard errors via statsmodels.

‚úÖ Option 3: Weighted Least Squares (WLS)
